```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
numpyDizi =np.linspace(0,10,10)
numpyDizi2 = numpyDizi ** 3
```

```python
figur = plt.figure(figsize=(10,10))
#dpi=150
eksen = figur.add_axes([0.1,0.1,0.9,0.9])
eksen.plot(numpyDizi,numpyDizi2 * 2,label = "numpyDizi ** 2")
eksen.plot(numpyDizi,numpyDizi2 * 3,label = "numpyDizi ** 3")
eksen.legend(loc=2) #çizgi açıklaması
```

```python
figur.savefig("Myfigure.png")  # Kaydetme
```
