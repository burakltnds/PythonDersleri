```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
numpyDizi =np.linspace(0,10,10)
numpyDizi
```

```python
numpyDizi2 = numpyDizi ** 3
```

```python
plt.plot(numpyDizi,numpyDizi2,"r*-")
plt.xlabel("Numpy1")
plt.ylabel("Numpy2")
plt.grid()
plt.show()
```

# Alt Grafik Üretme

```python
plt.subplot(1,2,1) # 1.si sıra 2.si kaç grafik olcagı 3.süde bunun kaçıncı grafik oldugu
plt.plot(numpyDizi,numpyDizi2 ,"r*-")
plt.grid()
plt.subplot(1,2,2) 
plt.plot(numpyDizi2,numpyDizi ,"g--")
plt.grid()
plt.show()
```
