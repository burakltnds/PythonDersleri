```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
numpyDizi =np.linspace(0,10,10)
numpyDizi2 = numpyDizi ** 2
```

```python
(figur , eksen) = plt.subplots()
eksen.plot(numpyDizi,numpyDizi2,"#00FF00",alpha=0.3 , linewidth = 8) # linewidth çizgi kalınlığı
eksen.plot(numpyDizi2,numpyDizi,"#8B008B",alpha=0.3 , linewidth=6) #alpha saydamlık
plt.show()
```
