```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
numpyDizi =np.linspace(0,10,10)
numpyDizi2 = numpyDizi ** 3
```

```python
figur = plt.figure()
figurAxes = figur.add_axes([0.2,0.2,0.9,0.9])
figurAxes.plot(numpyDizi,numpyDizi2,"g")
figurAxes.set_xlabel("X")
figurAxes.set_ylabel("Y")
```

# Devam

```python
figure = plt.figure()
eksen1 = figure.add_axes([0.1,0.1,0.9,0.9])
eksen2 = figure.add_axes([0.25,0.50,0.3,0.3])

eksen1.plot(numpyDizi,numpyDizi2,"b")
eksen1.set_xlabel("X")
eksen1.set_ylabel("Y")
eksen1.grid()

eksen2.plot(numpyDizi2,numpyDizi,"r")
eksen2.set_xlabel("X")
eksen2.set_ylabel("Y")
eksen2.grid()


```

# 2.KISIM

```python
(figure1 , eksen) = plt.subplots(nrows=1 , ncols=2)
for i in eksen:
    i.plot(numpyDizi , numpyDizi2,"g")
    i.set_xlabel("X")
plt.tight_layout()    
```
