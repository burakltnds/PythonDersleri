```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
numpyDizi =np.linspace(0,10,10)
numpyDizi2 = numpyDizi ** 2
```

# Scatter

```python
plt.scatter(numpyDizi,numpyDizi2)
```

# Histogram(artış)

```python
plt.hist(numpyDizi,numpyDizi2)
plt.show()
```

# Boxplot(standart sapma)

```python
plt.boxplot(numpyDizi2)
```
