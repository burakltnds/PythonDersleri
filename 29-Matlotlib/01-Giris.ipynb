```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
yas = [10,21,32,45,78,56,60,80,90,100]
kilo = [34,92,70,68,75,49,60,80,65,60]
```

```python
numpyYas = np.array(yas)
numpyKilo = np.array(kilo)
```

```python
numpyYas
numpyKilo
```

```python
plt.plot(numpyYas,numpyKilo,"red")
plt.title("Yaş-Kilo Grafiği")
plt.ylabel("Kilo")
plt.xlabel("Yaş")
plt.grid(True) #ızgara
plt.show()
```
